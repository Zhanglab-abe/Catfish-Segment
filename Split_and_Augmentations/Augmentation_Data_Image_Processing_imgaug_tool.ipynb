```python
import imageio
import imgaug as ia
from imgaug import augmenters as iaa
import numpy as np

maskisRGB = False



def dealImage(imagename,datapath,savepath):
    image = imageio.imread(f"{datapath}/images_raw/{imagename}.jpg")
    image1 = imageio.imread(f"{datapath}/annotations/{imagename}.png",pilmode="RGB")
    # image1 = image1 // 50
    print(np.unique(image1))
    ia.seed(4)  # 随机值

    images = [image]*5
    images1 = [image1]*5

    seq = iaa.Sequential(
        [
        iaa.Affine(rotate=(-25, 25),order=0,cval=1),                  # 旋转区间
        # iaa.AdditiveGaussianNoise(scale=(10, 60)),     # 添加高斯白噪声
        # iaa.Crop(percent=(0, 0.2)),                    # 实现裁剪，参数为裁剪比例
        # iaa.AddToHueAndSaturation((-60, 60)),          # Hue空间调整饱和度
        iaa.Fliplr(0.5),                               # 水平翻转
        iaa.Flipud(0.4),                               # 垂直翻转
        iaa.Resize({"height": "keep-aspect-ratio", "width": 640},interpolation=0),
        # iaa.Cutout(nb_iterations=(1, 2), size=0.2, squared=True),  # 图像增强之 cutout 生成遮挡块
        #   iaa.ElasticTransformation(alpha=90, sigma=9)  #随机弹性变换（仿射变换的一种）
        ],
        random_order=True)

    seq_det = seq.to_deterministic()                    # 每次调用seq_det使得batch的变换相同

    images_aug = seq_det(images=images)
    images_aug1 = seq_det(images=images1)

    # print(images_aug)

    for i in range(len(images_aug)):
        imageio.imwrite(f"{savepath}/images_raw/{imagename}_{i}.jpg",images_aug[i])
    for i in range(len(images_aug1)):
        print(images_aug1[i].shape)
        if maskisRGB:
            mask = images_aug1[i]
        else:
            mask = images_aug1[i].transpose()[0]
            mask = mask.transpose()
        print(np.unique(mask))
        imageio.imwrite(f"{savepath}/annotations/{imagename}_{i}.png",mask)

    # print("Augmented:")
    # # ia.imshow(np.hstack(images_aug))
    # ia.imshow(ia.draw_grid(images_aug, cols=4, rows=2))
    # ia.imshow(ia.draw_grid(images_aug1, cols=4, rows=2))
```
